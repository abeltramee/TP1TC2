### 1. Datos del Filtro B (Pasa Altos)
* **$f_p$ (Corte):** 4.6 kHz
* **$f_s$ (Stop):** 1.2 kHz
* **$\alpha_{max}$ (Ripple):** 1 dB
* **$\alpha_{min}$ (AtenuaciĂłn):** 20 dB

### 2. TransformaciĂłn al Prototipo Pasabajos
Para un filtro pasa altos, la frecuencia de stop normalizada ($\Omega_s$) se calcula mediante la relaciĂłn inversa de frecuencias:
$$\Omega_s = \frac{f_p}{f_s}$$

Sustituyendo los valores de la tabla:
$$\Omega_s = \frac{4.6 \text{ kHz}}{1.2 \text{ kHz}} \approx 3.833 \text{ rad/s}$$

* **$\Omega_p$ (Frecuencia de paso normalizada):** 1 rad/s

### 3. CĂĄlculo de $\epsilon^2$
A partir del ripple ($\alpha_{max} = 1 \text{ dB}$):
$$10 \cdot \log_{10}(1 + \epsilon^2) = 1 \text{ dB} \implies \epsilon^2 = 10^{0.1} - 1 = 0.2589$$

### 4. ParĂĄmetros para la iteraciĂłn
* **$\text{acosh}(\Omega_s)$:** $\text{acosh}(3.833) \approx 2.019$
* **$\alpha_{min}$ objetivo:** $\geq 20 \text{ dB}$

### 5. IteraciĂłn del orden ($n$)
FĂłrmula: $\alpha(dB) = 10 \cdot \log_{10}\left(1 + \epsilon^2 \cdot \cosh(n \cdot \text{acosh}(\Omega_s))^2\right)$

* **Para $n = 1$:**
  $$\alpha = 10 \cdot \log_{10}\left(1 + 0.2589 \cdot \cosh(1 \cdot 2.019)^2\right) \approx 6.81 \text{ dB} < 20 \text{ dB}$$
* **Para $n = 2$:**
  $$\alpha = 10 \cdot \log_{10}\left(1 + 0.2589 \cdot \cosh(2 \cdot 2.019)^2\right) \approx 23.2 \text{ dB} \geq 20 \text{ dB}$$

### ConclusiĂłn
* **Tipo:** Prototipo pasabajos de Chebyshev
* **Orden mĂ­nimo:** $N = 2$
* **Ripple ($\alpha_{max}$):** 1 dB
* **$\Omega_p$:** 1 rad/s

### 6. CĂĄlculo del parĂĄmetro auxiliar ($a$)
Tenemos los siguientes datos del paso anterior:
* **$N = 2$**
* **$\epsilon^2 = 0.2589 \implies \epsilon = \sqrt{0.2589} \approx 0.5088$**

Aplicamos la fĂłrmula para $a$:
$$a = \frac{1}{2} \cdot \text{asinh}\left(\frac{1}{0.5088}\right)$$
$$a = 0.5 \cdot \text{asinh}(1.9654)$$
$$a \approx 0.5 \cdot 1.4276 \approx 0.7138$$

Con este valor de $a$, calculamos las funciones hiperbĂłlicas:
* **$\sinh(a) = \sinh(0.7138) \approx 0.7765$**
* **$\cosh(a) = \cosh(0.7138) \approx 1.2662$**

### 7. CĂĄlculo de los Polos ($k = 1$ y $k = 2$)
Evaluamos para $N=2$. El tĂŠrmino del ĂĄngulo es $\theta_k = \frac{(2k-1)\pi}{4}$.

**Para $k = 1$:**
* Ăngulo: $\theta_1 = \frac{\pi}{4} \text{ rad} = 45^\circ$
* $\sin(45^\circ) = \cos(45^\circ) \approx 0.7071$
* **$\sigma_1$** $= -0.7765 \cdot 0.7071 \approx -0.5491$
* **$\omega_1$** $= 1.2662 \cdot 0.7071 \approx 0.8953$
* **Polo 1:** $p_1 = -0.5491 + j0.8953$

**Para $k = 2$:**
* Ăngulo: $\theta_2 = \frac{3\pi}{4} \text{ rad} = 135^\circ$
* $\sin(135^\circ) \approx 0.7071$
* $\cos(135^\circ) \approx -0.7071$
* **$\sigma_2$** $= -0.7765 \cdot 0.7071 \approx -0.5491$
* **$\omega_2$** $= 1.2662 \cdot (-0.7071) \approx -0.8953$
* **Polo 2:** $p_2 = -0.5491 - j0.8953$

### 8. FunciĂłn de Transferencia del Prototipo Pasabajos ($H_{LP}(s)$)
El denominador $D(s)$ se forma con el producto de las raĂ­ces: $(s - p_1)(s - p_2)$
$$D(s) = s^2 - (p_1 + p_2)s + (p_1 \cdot p_2)$$
* **$p_1 + p_2$** $= -1.0982$
* **$p_1 \cdot p_2$** $= (-0.5491)^2 + (0.8953)^2 = 0.3015 + 0.8016 = 1.1031$

Por lo tanto:
$$D(s) = s^2 + 1.0982s + 1.1031$$

Para filtros de Chebyshev de **orden par**, la ganancia en DC (numerador $K$) se ajusta para que el filtro arranque en $-\alpha_{max}$ (es decir, atenuado por el ripple):
$$K = \frac{p_1 \cdot p_2}{\sqrt{1 + \epsilon^2}} = \frac{1.1031}{\sqrt{1 + 0.2589}} = \frac{1.1031}{1.122} \approx 0.9831$$

**FunciĂłn de Transferencia Final del Prototipo:**
$$H_{LP}(s) = \frac{0.9831}{s^2 + 1.0982s + 1.1031}$$

### 9. EcuaciĂłn de TransformaciĂłn
Para convertir un prototipo pasabajos normalizado a un filtro pasa altos, se aplica la sustituciĂłn:
$$s = \frac{\omega_c}{S}$$

Donde $\omega_c$ es la frecuencia angular de corte del pasa altos:
$$\omega_c = 2 \pi f_p = 2 \pi \cdot 4600 \approx 28902.65 \text{ rad/s}$$

Por lo tanto:
$$s = \frac{28902.65}{S}$$

### 10. SustituciĂłn en $H_{LP}(s)$
Partimos de la funciĂłn de transferencia del pasabajos:
$$H_{LP}(s) = \frac{0.9831}{s^2 + 1.0982s + 1.1031}$$

Reemplazamos $s$:
$$H_{HP}(S) = \frac{0.9831}{\left(\frac{28902.65}{S}\right)^2 + 1.0982 \cdot \left(\frac{28902.65}{S}\right) + 1.1031}$$

### 11. Desarrollo Algebraico
Multiplicamos numerador y denominador por $S^2$:
$$H_{HP}(S) = \frac{0.9831 \cdot S^2}{(28902.65)^2 + 1.0982 \cdot (28902.65) \cdot S + 1.1031 \cdot S^2}$$

Calculamos las constantes:
* $(28902.65)^2 \approx 835363177$
* $1.0982 \cdot 28902.65 \approx 31740.89$

Reordenamos el denominador:
$$H_{HP}(S) = \frac{0.9831 \cdot S^2}{1.1031 \cdot S^2 + 31740.89 \cdot S + 835363177}$$

### 12. NormalizaciĂłn de $H_{HP}(S)$
Para dejar el coeficiente de $S^2$ del denominador en 1, dividimos todos los tĂŠrminos por **1.1031**:
* **Numerador:** $\frac{0.9831}{1.1031} \approx 0.8912$
* **TĂŠrmino en $S$:** $\frac{31740.89}{1.1031} \approx 28774.26$
* **TĂŠrmino independiente:** $\frac{835363177}{1.1031} \approx 757286897.8$

### 13. FunciĂłn de Transferencia Final (Pasa Altos)
$$H_{HP}(S) = \frac{0.8912 \cdot S^2}{S^2 + 28774.26 \cdot S + 757286897.8}$$
