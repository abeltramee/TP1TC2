### 1. Datos del Filtro B (Pasa Altos)
* **$f_p$ (Corte):** 4.6 kHz
* **$f_s$ (Stop):** 1.2 kHz
* **$\alpha_{max}$ (Ripple):** 1 dB
* **$\alpha_{min}$ (Atenuación):** 20 dB

### 2. Transformación al Prototipo Pasabajos
Para un filtro pasa altos, la frecuencia de stop normalizada ($\Omega_s$) se calcula mediante la relación inversa de frecuencias:
$$\Omega_s = \frac{f_p}{f_s}$$

Sustituyendo los valores de la tabla:
$$\Omega_s = \frac{4.6 \text{ kHz}}{1.2 \text{ kHz}} \approx 3.833 \text{ rad/s}$$

* **$\Omega_p$ (Frecuencia de paso normalizada):** 1 rad/s

### 3. Cálculo de $\epsilon^2$
A partir del ripple ($\alpha_{max} = 1 \text{ dB}$):
$$10 \cdot \log_{10}(1 + \epsilon^2) = 1 \text{ dB} \implies \epsilon^2 = 10^{0.1} - 1 = 0.2589$$

### 4. Parámetros para la iteración
* **$\text{acosh}(\Omega_s)$:** $\text{acosh}(3.833) \approx 2.019$
* **$\alpha_{min}$ objetivo:** $\geq 20 \text{ dB}$

### 5. Iteración del orden ($n$)
Fórmula: $\alpha(dB) = 10 \cdot \log_{10}\left(1 + \epsilon^2 \cdot \cosh(n \cdot \text{acosh}(\Omega_s))^2\right)$

* **Para $n = 1$:**
  $$\alpha = 10 \cdot \log_{10}\left(1 + 0.2589 \cdot \cosh(1 \cdot 2.019)^2\right) \approx 6.81 \text{ dB} < 20 \text{ dB}$$
* **Para $n = 2$:**
  $$\alpha = 10 \cdot \log_{10}\left(1 + 0.2589 \cdot \cosh(2 \cdot 2.019)^2\right) \approx 23.2 \text{ dB} \geq 20 \text{ dB}$$

### Conclusión
* **Tipo:** Prototipo pasabajos de Chebyshev
* **Orden mínimo:** $N = 2$
* **Ripple ($\alpha_{max}$):** 1 dB
* **$\Omega_p$:** 1 rad/s

### 6. Cálculo del parámetro auxiliar ($a$)
Tenemos los siguientes datos del paso anterior:
* **$N = 2$**
* **$\epsilon^2 = 0.2589 \implies \epsilon = \sqrt{0.2589} \approx 0.5088$**

Aplicamos la fórmula para $a$:
$$a = \frac{1}{2} \cdot \text{asinh}\left(\frac{1}{0.5088}\right)$$
$$a = 0.5 \cdot \text{asinh}(1.9654)$$
$$a \approx 0.5 \cdot 1.4276 \approx 0.7138$$

Con este valor de $a$, calculamos las funciones hiperbólicas:
* **$\sinh(a) = \sinh(0.7138) \approx 0.7765$**
* **$\cosh(a) = \cosh(0.7138) \approx 1.2662$**

### 7. Cálculo de los Polos ($k = 1$ y $k = 2$)
Evaluamos para $N=2$. El término del ángulo es $\theta_k = \frac{(2k-1)\pi}{4}$.

**Para $k = 1$:**
* Ángulo: $\theta_1 = \frac{\pi}{4} \text{ rad} = 45^\circ$
* $\sin(45^\circ) = \cos(45^\circ) \approx 0.7071$
* **$\sigma_1$** $= -0.7765 \cdot 0.7071 \approx -0.5491$
* **$\omega_1$** $= 1.2662 \cdot 0.7071 \approx 0.8953$
* **Polo 1:** $p_1 = -0.5491 + j0.8953$

**Para $k = 2$:**
* Ángulo: $\theta_2 = \frac{3\pi}{4} \text{ rad} = 135^\circ$
* $\sin(135^\circ) \approx 0.7071$
* $\cos(135^\circ) \approx -0.7071$
* **$\sigma_2$** $= -0.7765 \cdot 0.7071 \approx -0.5491$
* **$\omega_2$** $= 1.2662 \cdot (-0.7071) \approx -0.8953$
* **Polo 2:** $p_2 = -0.5491 - j0.8953$

### 8. Función de Transferencia del Prototipo Pasabajos ($H_{LP}(s)$)
El denominador $D(s)$ se forma con el producto de las raíces: $(s - p_1)(s - p_2)$
$$D(s) = s^2 - (p_1 + p_2)s + (p_1 \cdot p_2)$$
* **$p_1 + p_2$** $= -1.0982$
* **$p_1 \cdot p_2$** $= (-0.5491)^2 + (0.8953)^2 = 0.3015 + 0.8016 = 1.1031$

Por lo tanto:
$$D(s) = s^2 + 1.0982s + 1.1031$$

Para filtros de Chebyshev de **orden par**, la ganancia en DC (numerador $K$) se ajusta para que el filtro arranque en $-\alpha_{max}$ (es decir, atenuado por el ripple):
$$K = \frac{p_1 \cdot p_2}{\sqrt{1 + \epsilon^2}} = \frac{1.1031}{\sqrt{1 + 0.2589}} = \frac{1.1031}{1.122} \approx 0.9831$$

**Función de Transferencia Final del Prototipo:**
$$H_{LP}(s) = \frac{0.9831}{s^2 + 1.0982s + 1.1031}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 1. Definir la función de transferencia calculada H(s)
# H(s) = 0.9831 / (s^2 + 1.0982s + 1.1031)
num = [0.9831]
den = [1, 1.0982, 1.1031]

# Crear el sistema LTI
sys = signal.TransferFunction(num, den)

# 2. Generar vector de frecuencias angulares (w) en rad/s
w = np.logspace(-1, 1, 1000)

# 3. Calcular la respuesta en frecuencia (Bode)
w, mag, phase = signal.bode(sys, w)

# 4. Graficar
plt.figure(figsize=(10, 6))
plt.semilogx(w, mag, label='H(s) Calculada', linewidth=2)

# --- Líneas de Verificación de Especificaciones ---

# Banda de paso: Corte en 1 rad/s, atenuación máxima -1 dB
plt.axvline(x=1, color='green', linestyle='--', label=r'$\Omega_p = 1$ rad/s')
plt.axhline(y=-1, color='red', linestyle=':', label=r'$\alpha_{max} = -1$ dB')

# Banda de stop: Stop en 3.833 rad/s, atenuación mínima -20 dB
plt.axvline(x=3.833, color='orange', linestyle='--', label=r'$\Omega_s = 3.833$ rad/s')
plt.axhline(y=-20, color='purple', linestyle=':', label=r'$\alpha_{min} = -20$ dB')

# Configuraciones del gráfico
plt.title('Verificación del Prototipo Pasabajos Chebyshev (N=2)')
plt.xlabel('Frecuencia Angular $\Omega$ [rad/s]')
plt.ylabel('Magnitud [dB]')
plt.grid(which='both', axis='both', linestyle='--', alpha=0.7)
plt.legend(loc='lower left')

# Ajustar límites de los ejes para mejor visualización
plt.xlim(0.1, 10)
plt.ylim(-40, 5)

plt.show()
```
